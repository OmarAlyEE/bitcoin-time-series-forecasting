```python
from pathlib import Path
import pandas as pd

# Use the current working directory of the notebook
BASE_DIR = Path().resolve().parents[0]  # adjust number of `.parents` as needed
preds_file = BASE_DIR / "data/processed/predictions_improved.csv"

# Load predictions
preds_df = pd.read_csv(preds_file)
preds_df['date'] = pd.to_datetime(preds_df['date'])
```

```python
plt.figure(figsize=(15,6))
plt.plot(preds_df['date'], preds_df['actual'], label='Actual', color='black', lw=1.2)
plt.plot(preds_df['date'], preds_df['gru'], label='Multivariate GRU', color='#1f77b4', lw=1.8, alpha=0.9)
plt.plot(preds_df['date'], preds_df['ridge'], label='Ridge Regression', color='#2ca02c', lw=1.6, alpha=0.75)
plt.plot(preds_df['date'], preds_df['ensemble'], label='Ensemble (GRU + Ridge)', color='#d62728', lw=2.3)

plt.title("Bitcoin Price Prediction — Test Set", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend(frameon=True, shadow=True)
plt.grid(True, ls="--", alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
recent = preds_df.tail(60)

plt.figure(figsize=(14,5))
plt.plot(recent['date'], recent['actual'], label='Actual', color='black', lw=1.2)
plt.plot(recent['date'], recent['ridge'], label='Ridge Regression', color='green', lw=1.8)
plt.title("Bitcoin Price Prediction — Last 60 Days")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True, ls="--", alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
preds_df['ridge_error'] = preds_df['ridge'] - preds_df['actual']

plt.figure(figsize=(14,5))
plt.plot(preds_df['date'], preds_df['ridge_error'], label='Ridge Prediction Error', color='red')
plt.title("Ridge Prediction Error Over Time")
plt.xlabel("Date")
plt.ylabel("Prediction Error (USD)")
plt.grid(True, ls="--", alpha=0.3)
plt.tight_layout()
plt.show()
```
